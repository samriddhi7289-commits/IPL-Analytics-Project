# Question 1: Do Teams That Win the Toss Actually Win More Matches?

The toss plays an important role in T20 cricket because teams can choose their strategy depending on pitch conditions, dew factor, and match pressure.

This analysis investigates whether winning the toss actually increases the chances of winning IPL matches.


```python
# =========================================================
# QUESTION 1 — DO TEAMS THAT WIN THE TOSS WIN MORE MATCHES?
# =========================================================

match_data = ipl.drop_duplicates(subset='match_id')

toss_match_win = match_data[
    match_data['toss_winner'] == match_data['match_won_by']
]

won_after_toss = toss_match_win.shape[0]

total_matches = match_data.shape[0]

lost_after_toss = total_matches - won_after_toss

# Percentages

won_percent = round((won_after_toss / total_matches) * 100, 1)

lost_percent = round((lost_after_toss / total_matches) * 100, 1)

labels = ['Won Match', 'Lost Match']

values = [won_after_toss, lost_after_toss]

plt.figure(figsize=(8,5))

bars = plt.bar(
    labels,
    values,
    color=['#00B4D8', '#FF6B35'],
    width=0.8
)

plt.title(
    'Impact of Toss on IPL Match Results',
    fontsize=18,
    fontweight='bold'
)

plt.ylabel('Number of Matches', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.2)

for bar in bars:

    yval = bar.get_height()

    percent = round((yval / total_matches) * 100, 1)

    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval + 5,
        f'{int(yval)}\n({percent}%)',
        ha='center',
        fontsize=11,
        fontweight='bold'
    )

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The analysis shows that winning the toss provides some advantage, but it does not guarantee victory. Match performance, momentum, and execution during crucial overs still play a larger role in determining IPL outcomes.


# Question 2: Has IPL Become Batter-Dominated Over Time?

Over the years, T20 cricket has evolved significantly with more aggressive batting approaches, innovative shot-making, and higher scoring rates.

This analysis explores whether IPL has gradually become more batter-friendly by examining season-wise scoring trends from 2008 to 2025.


```python
# =========================================================
# QUESTION 2 — HAS IPL BECOME BATTER-DOMINATED?
# =========================================================

match_runs = ipl.groupby(
    ['match_id', 'season']
)['runs_total'].sum().reset_index()

season_avg = match_runs.groupby(
    'season'
)['runs_total'].mean().reset_index()

# Season Cleaning

season_avg['season'] = season_avg['season'].replace({
    '2007/08': '2008',
    '2009/10': '2010',
    '2020/21': '2020'
})

season_avg['season'] = season_avg['season'].astype(str)

# Remove 2026

season_avg = season_avg[
    season_avg['season'] != '2026'
]

# Remove Duplicates

season_avg = season_avg.drop_duplicates(
    subset='season'
)

# Numerical Sorting

season_avg['season_num'] = season_avg['season'].astype(int)

season_avg = season_avg.sort_values(
    'season_num'
)

# Plot

plt.figure(figsize=(14,6))

plt.plot(
    season_avg['season'],
    season_avg['runs_total'],
    marker='o',
    linewidth=3,
    markersize=7,
    color='#0077B6'
)

plt.title(
    'Has IPL Become Batter-Dominated Over Time?',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Season', fontsize=12)

plt.ylabel('Average Match Runs', fontsize=12)

plt.grid(
    linestyle='--',
    alpha=0.2
)

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The graph clearly shows a gradual increase in IPL scoring trends over the years. Modern T20 cricket has become increasingly batter-friendly due to aggressive batting strategies, improved strike rates, and evolving match tactics.


# Question 3: Which Phase Impacts Victory the Most?

Different phases of a T20 match play different tactical roles. Powerplays encourage aggressive batting, middle overs focus on stability and strike rotation, while death overs often determine the final momentum of the game.

This analysis examines which phase contributes the most towards match outcomes by comparing scoring patterns across IPL seasons.


```python
# =========================================================
# QUESTION 3 — WHICH PHASE IMPACTS VICTORY THE MOST?
# =========================================================

def phase(over):

    if over <= 6:
        return 'Powerplay'

    elif over <= 15:
        return 'Middle Overs'

    else:
        return 'Death Overs'

ipl['phase'] = ipl['over'].apply(phase)

phase_runs = ipl.groupby(
    'phase'
)['runs_total'].mean().reset_index()

phase_order = [
    'Powerplay',
    'Middle Overs',
    'Death Overs'
]

phase_runs['phase'] = pd.Categorical(
    phase_runs['phase'],
    categories=phase_order,
    ordered=True
)

phase_runs = phase_runs.sort_values('phase')

plt.figure(figsize=(10,5))

bars = plt.bar(
    phase_runs['phase'],
    phase_runs['runs_total'],
    color=['#00B4D8', '#F4A261', '#E63946'],
    width=0.6
)

plt.title(
    'Average Runs Across Match Phases',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Match Phase', fontsize=12)

plt.ylabel('Average Runs Per Ball', fontsize=12)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.2
)

for bar in bars:

    yval = round(bar.get_height(), 2)

    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval + 0.02,
        yval,
        ha='center',
        fontsize=11,
        fontweight='bold'
    )

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The analysis highlights how scoring intensity changes throughout the innings. Death overs generally produce the highest scoring rates because teams adopt aggressive batting strategies during the final overs to maximize totals or accelerate run chases.


# Question 4: Which Overs Create Match-Winning Momentum?

Momentum plays a crucial role in T20 cricket. Certain overs can dramatically change the direction of a match by increasing scoring rates or creating pressure through wickets.

This analysis identifies which overs contribute the most towards match-winning momentum by examining average runs scored across overs.


```python
# =========================================================
# QUESTION 4 — WHICH OVERS CREATE MATCH-WINNING MOMENTUM?
# =========================================================

over_runs = ipl.groupby(
    'over'
)['runs_total'].mean().reset_index()

over_runs = over_runs[
    over_runs['over'] <= 20
]

plt.figure(figsize=(14,6))

plt.plot(
    over_runs['over'],
    over_runs['runs_total'],
    marker='o',
    linewidth=3,
    markersize=6,
    color='#4361EE'
)

# Highlight Death Overs

plt.axvspan(
    16,
    20,
    alpha=0.08,
    color='red'
)

# Vertical Line

plt.axvline(
    x=16,
    linestyle='--',
    alpha=0.5,
    color='red'
)

plt.title(
    'Average Runs Across Overs',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Overs', fontsize=12)

plt.ylabel('Average Runs Per Ball', fontsize=12)

plt.grid(
    linestyle='--',
    alpha=0.2
)

plt.xticks(range(1,21))

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The graph shows that scoring rates increase significantly during the death overs. Overs 16–20 often create the strongest momentum shifts because teams accelerate aggressively to maximize totals or chase targets under pressure.


# Question 5: Can We Predict Collapses Before They Happen?

Batting collapses are one of the most dramatic events in T20 cricket. Consecutive wickets and sudden drops in scoring rates can completely change the outcome of a match.

This analysis studies wicket patterns across overs to identify when batting collapses are most likely to occur in IPL matches.


```python
# =========================================================
# QUESTION 5 — CAN WE PREDICT COLLAPSES?
# =========================================================

wickets_over = ipl.groupby(
    'over'
)['bowler_wicket'].sum().reset_index()

wickets_over = wickets_over[
    wickets_over['over'] <= 20
]

plt.figure(figsize=(14,6))

plt.plot(
    wickets_over['over'],
    wickets_over['bowler_wicket'],
    marker='o',
    linewidth=3,
    markersize=6,
    color='#D62828'
)

plt.axvspan(
    15,
    20,
    alpha=0.08,
    color='orange'
)

plt.title(
    'Wickets Falling Across Overs',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Overs', fontsize=12)

plt.ylabel('Total Wickets', fontsize=12)

plt.grid(
    linestyle='--',
    alpha=0.2
)

plt.xticks(range(1,21))

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The analysis shows that wicket frequency increases significantly during later overs when teams attempt aggressive acceleration. Pressure situations during death overs often trigger batting collapses and rapid momentum shifts.


# Question 6: What Truly Predicts Winning in IPL?

Many factors influence IPL victories, including powerplay performance, death-over execution, venue conditions, and batting totals.

This analysis focuses on whether higher first-innings scores significantly improve the chances of winning matches.


```python
# =========================================================
# QUESTION 6 — WHAT TRULY PREDICTS WINNING?
# =========================================================

first_innings = ipl[
    ipl['innings'] == 1
]

first_innings_score = first_innings.groupby(
    ['match_id', 'batting_team']
)['runs_total'].sum().reset_index()

first_innings_score.rename(
    columns={'runs_total': 'first_innings_runs'},
    inplace=True
)

match_winner = ipl.drop_duplicates(
    subset='match_id'
)[['match_id', 'match_won_by']]

winning_data = pd.merge(
    first_innings_score,
    match_winner,
    on='match_id'
)

winning_data['won_match'] = np.where(
    winning_data['batting_team'] ==
    winning_data['match_won_by'],
    1,
    0
)

bins = [0, 140, 160, 180, 200, 250]

labels = [
    'Below 140',
    '140-160',
    '160-180',
    '180-200',
    '200+'
]

winning_data['score_range'] = pd.cut(
    winning_data['first_innings_runs'],
    bins=bins,
    labels=labels
)

win_percent = winning_data.groupby(
    'score_range'
)['won_match'].mean() * 100

win_percent = win_percent.reset_index()

plt.figure(figsize=(10,5))

bars = plt.bar(
    win_percent['score_range'],
    win_percent['won_match'],
    color='#0077B6',
    width=0.6
)

plt.title(
    'Winning Percentage by First Innings Score',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    'First Innings Score Range',
    fontsize=12
)

plt.ylabel(
    'Winning Percentage',
    fontsize=12
)

plt.grid(
    axis='y',
    linestyle='--',
    alpha=0.2
)

for bar in bars:

    yval = round(bar.get_height(), 1)

    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval + 1,
        f'{yval}%',
        ha='center',
        fontsize=11,
        fontweight='bold'
    )

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The scatter plot suggests that higher first-innings scores generally improve winning probability. However, several matches are still lost despite strong totals, showing that momentum shifts, bowling execution, and chase strategies remain equally important in IPL cricket.


# Question 7: Is Aggressive Powerplay Batting Worth the Risk?

Powerplay overs are considered one of the most aggressive phases in T20 cricket. Teams often attack heavily during the first six overs to gain an early advantage.

This analysis studies scoring patterns during powerplays to understand whether aggressive starts contribute significantly towards match success.


```python
# =========================================================
# QUESTION 7 — IS AGGRESSIVE POWERPLAY BATTING WORTH IT?
# =========================================================

powerplay = ipl[
    ipl['over'] <= 6
]

pp_runs = powerplay.groupby(
    'over'
)['runs_total'].mean().reset_index()

plt.figure(figsize=(12,5))

plt.plot(
    pp_runs['over'],
    pp_runs['runs_total'],
    marker='o',
    linewidth=3,
    markersize=7,
    color='#2A9D8F'
)

plt.axvspan(
    1,
    6,
    alpha=0.08,
    color='#2A9D8F'
)

plt.title(
    'Average Runs During Powerplay Overs',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Overs', fontsize=12)

plt.ylabel('Average Runs Per Ball', fontsize=12)

plt.grid(
    linestyle='--',
    alpha=0.2
)

plt.xticks(range(1,7))

plt.xlim(1,6)

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The graph highlights how teams adopt aggressive batting strategies during powerplay overs to maximize scoring opportunities while field restrictions are active. Strong powerplay starts often create early momentum and reduce scoreboard pressure later in the innings.


# Question 8: Does Venue Have More Impact Than Team Strength?

Venue conditions can heavily influence T20 matches. Pitch behavior, boundary dimensions, weather, and dew conditions often affect scoring patterns and match outcomes.

This analysis explores how different IPL venues impact scoring trends and whether venue conditions influence matches more than team strength alone.


```python
# =========================================================
# QUESTION 8 — DOES VENUE IMPACT MATCHES?
# =========================================================

ipl['venue'] = ipl['venue'].replace({

    'Maharaja Yadavindra Singh International Cricket Stadium, New Chandigarh':
    'New Chandigarh Stadium',

    'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium, Visakhapatnam':
    'ACA-VDCA Stadium',

    'Himachal Pradesh Cricket Association Stadium, Dharamsala':
    'Dharamsala Stadium',

    'Punjab Cricket Association IS Bindra Stadium, Mohali, Chandigarh':
    'Mohali Stadium',

    'Rajiv Gandhi International Stadium, Uppal, Hyderabad':
    'Hyderabad Stadium',

    'M Chinnaswamy Stadium, Bengaluru':
    'Chinnaswamy Stadium'
})

venue_runs = ipl.groupby(
    'venue'
)['runs_total'].mean().reset_index()

venue_runs = venue_runs.sort_values(
    by='runs_total',
    ascending=False
).head(10)

venue_runs = venue_runs[::-1]

plt.figure(figsize=(12,6))

bars = plt.barh(
    venue_runs['venue'],
    venue_runs['runs_total'],
    color='#9D4EDD'
)

plt.title(
    'Top IPL Venues by Average Runs',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    'Average Runs Per Ball',
    fontsize=12
)

plt.ylabel(
    'Venue',
    fontsize=12
)

plt.grid(
    axis='x',
    linestyle='--',
    alpha=0.2
)

for bar in bars:

    xval = round(bar.get_width(), 2)

    plt.text(
        xval + 0.02,
        bar.get_y() + bar.get_height()/2,
        xval,
        va='center',
        fontsize=10,
        fontweight='bold'
    )

sns.despine()

plt.tight_layout()

plt.show()
```

### Insight

The analysis shows that venue conditions significantly influence IPL scoring patterns. Certain stadiums consistently produce higher-scoring matches due to pitch behavior, shorter boundaries, and batting-friendly conditions.
